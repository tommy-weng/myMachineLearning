```python
import matplotlib.pyplot as plt

# 支持中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 数据
sizes = [20, 20, 20, 20, 20] # 每个部分的大小（百分比）
labels = ['德', '智', '体', '美', '劳'] # 每个部分对应的标签
colors = ['red', 'green', 'blue', 'purple', 'orange'] # 每个部分对应的颜色
 
# 创建饼状图
plt.pie(sizes, labels=labels, colors=colors, startangle=90, #explode=(0.1,0.1,0.1), shadow=True, autopct='%.1f%%', 
    labeldistance=0.85, pctdistance=0.85, rotatelabels=True, 
    wedgeprops={'width':0.2, 'edgecolor':'w', 'linewidth':0})

plt.pie(sizes, colors=colors, startangle=90,
    labeldistance=1.2, pctdistance=0.85,
    wedgeprops={'width':0.2, 'edgecolor':'w', 'linewidth':3},
    radius=0.8)

plt.pie(sizes, colors=colors, startangle=90,
    labeldistance=1.2, pctdistance=0.85,
    wedgeprops={'width':0.2, 'edgecolor':'w', 'linewidth':3},
    radius=0.6)

plt.pie(sizes, colors=colors, startangle=90,
    labeldistance=1.2, pctdistance=0.85,
    wedgeprops={'width':0.2, 'edgecolor':'w', 'linewidth':3},
    radius=0.4)

# 添加标题
plt.title('Pie Chart')
 
# 显示图形
plt.show()
```
